```python
import os
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('./data/华泰证券.xlsx')
dst_col = 'close'
```

```python
df.head()
```

```python
df.sort_values(by='trade_date',ascending=True,inplace=True)
dst_data = df[dst_col].astype('float32').values
```

## 特征加工

```python
# 将数据集拆分为训练集和测试集（80/20）
training_data_len = int(np.ceil(len(dst_data) * 0.8))
train_data = dst_data[0:training_data_len]
test_data = dst_data[training_data_len - 2:]

# 将数据集缩放到0到1之间
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_train_data = scaler.fit_transform(train_data.reshape(-1,1))
scaled_test_data = scaler.transform(test_data.reshape(-1,1))

# 创建训练数据集和标签
def create_dataset(dataset, time_step=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - time_step):
        a = dataset[i:(i + time_step), 0]
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)


time_step = 2
X_train, y_train = create_dataset(scaled_train_data, time_step)
X_test, y_test = create_dataset(scaled_test_data, time_step)

# 将数据集重塑为LSTM所需的格式：[样本数，时间步数，特征数]
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# 用于最终绘图
all_data = np.concatenate((scaled_train_data,scaled_test_data))
all_data = np.reshape(all_data, (all_data.shape[0], all_data.shape[1], 1))
```

## rnn

```python
from tensorflow.keras.layers import Dense, SimpleRNN
# 创建RNN模型
model_rnn = Sequential()
model_rnn.add(SimpleRNN(50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model_rnn.add(SimpleRNN(50, return_sequences=False))
model_rnn.add(Dense(25))
model_rnn.add(Dense(1))

# 编译模型
model_rnn.compile(optimizer='adam', loss='mean_squared_error')

# 训练模型
model_rnn.fit(X_train, y_train, epochs=10, batch_size=8)

# 使用模型进行预测
predictions_rnn = model_rnn.predict(X_test)
predictions_rnn = scaler.inverse_transform(predictions_rnn)

# 计算模型指标
rmse = np.sqrt(np.mean(((predictions_rnn - test_data) ** 2)))
print('Root Mean Squared Error:', rmse)
```

```python
# 使用模型进行预测
predictions_rnn = model_rnn.predict(all_data)
predictions_rnn = scaler.inverse_transform(predictions_rnn)

predictions_rnn_test = model_rnn.predict(X_test)
predictions_rnn_test = scaler.inverse_transform(predictions_rnn_test)
```

```python
# y = list(range(1993,2023,1))
all_y = list(range(all_data.shape[0]))
tag = [i*250 for i in list(range(1,31,1))]
test_Rnn_tag = len(predictions_rnn_test)*-1

plt.figure(figsize=(16,9))
plt.plot(all_y,predictions_rnn, 'y', label='RNN')
plt.plot(dst_data, 'b', label='real')
plt.plot((training_data_len, training_data_len), (0, 1), 'g--')  # 分割线 左边是训练数据 右边是测试数据的输出
plt.legend(loc='best')
# plt.xticks(tag,y)
plt.xlabel("time")
plt.ylabel("mount")
plt.show()
```

## lstm

```python
# 创建LSTM模型
model_lstm = Sequential()
model_lstm.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model_lstm.add(LSTM(50, return_sequences=False))
model_lstm.add(Dense(25))
model_lstm.add(Dense(1))

# 编译模型
model_lstm.compile(optimizer='adam', loss='mean_squared_error')

# 训练模型
model_lstm.fit(X_train, y_train, epochs=10, batch_size=8)

# 使用模型进行预测
predictions_lstm = model_lstm.predict(X_test)
predictions_lstm = scaler.inverse_transform(predictions_lstm)

# 计算模型指标
rmse = np.sqrt(np.mean(((predictions_lstm - test_data) ** 2)))
print('Root Mean Squared Error:', rmse)
```

```python
# 使用模型进行预测
predictions_lstm = model_lstm.predict(all_data)
predictions_lstm = scaler.inverse_transform(predictions_lstm)

predictions_lstm_test = model_rnn.predict(X_test)
predictions_lstm_test = scaler.inverse_transform(predictions_lstm_test)
```

```python
plt.figure(figsize=(16,9))
plt.plot(all_y,predictions_rnn, 'y', label='RNN')
plt.plot(all_y,predictions_lstm, 'r', label='LSTM')
plt.plot(dst_col, 'b', label='real')
plt.plot((training_data_len, training_data_len), (0, 1), 'g--')  # 分割线 左边是训练数据 右边是测试数据的输出
plt.legend(loc='best')
# plt.xticks(tag,y)
plt.xlabel("time")
plt.ylabel("mount")
plt.show()
```

## dnn

```python
import tensorflow as tf

# 定义模型
model_dnn = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(time_step,)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)
])

# 编译模型
model_dnn.compile(optimizer='adam', loss='mse')

# 训练模型
model_dnn.fit(X_train, y_train, epochs=10, batch_size=32)

# # 评估模型
# mse = model_dnn.evaluate(X_test, y_test)
# print("Mean Squared Error:", mse)

# 使用模型进行预测
predictions_dnn = model_dnn.predict(X_test)
predictions_dnn = scaler.inverse_transform(predictions_dnn)

# 计算模型指标
rmse = np.sqrt(np.mean(((predictions_dnn - test_data) ** 2)))
print('Root Mean Squared Error:', rmse)
```

```python
all_data_dnn = np.concatenate((X_train,X_test))
predictions_dnn=model_dnn.predict(all_data_dnn)
predictions_dnn = scaler.inverse_transform(predictions_dnn)

predictions_dnn_test = model_dnn.predict(X_test)
predictions_dnn_test = scaler.inverse_transform(predictions_dnn_test)
```
